```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# drop irrelevant columns
df = merged_df
df = df.drop(['year', 'UNITID', 'INSTNM'], axis=1)

# get rid of privacy stuff
df.replace('PrivacySuppressed', np.nan, inplace=True)
df = df.dropna()

# OneHOt encoding based on major
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_features = encoder.fit_transform(df[['CIPDESC']])
categories = encoder.get_feature_names_out(['CIPDESC'])
df_encoded = pd.concat([df, pd.DataFrame(encoded_features, columns=categories)], axis=1)
df_encoded = df_encoded.drop('CIPDESC', axis=1)

# split by features (salaries)
X = df_encoded.drop(['EARN_MDN_4YR'], axis=1)
y = df_encoded['EARN_MDN_4YR']
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# train linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

#predict salaries
y_pred = model.predict(X_test)

#measure of model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# prediction output
new_data = pd.DataFrame({
    'CIPDESC': ['Computer and Information Sciences, General.'],  #major
    'CIPCODE': [1101],  #major code
    'CREDLEV': [5],  #degree needed
    'EARN_MDN_1YR': [np.nan],  
})

new_data_encoded = pd.concat([new_data, pd.DataFrame(encoder.transform(new_data[['CIPDESC']]), columns=categories)], axis=1)
new_data_encoded = new_data_encoded.drop('CIPDESC', axis=1)
new_data_imputed = pd.DataFrame(imputer.transform(new_data_encoded), columns=new_data_encoded.columns)

#predict using the trained model for the new data
new_predictions = model.predict(new_data_imputed)

print('Predicted Earnings 4 Year After Graduation:')
print(new_predictions)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from prettytable import PrettyTable

# drop irrelevant columns
df = merged_df
df = df.drop(['year', 'UNITID', 'INSTNM'], axis=1)

# get rid of privacy stuff
df.replace('PrivacySuppressed', np.nan, inplace=True)
df = df.dropna()

# OneHOt encoding based on major
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_features = encoder.fit_transform(df[['CIPDESC']])
categories = encoder.get_feature_names_out(['CIPDESC'])
df_encoded = pd.concat([df, pd.DataFrame(encoded_features, columns=categories)], axis=1)
df_encoded = df_encoded.drop('CIPDESC', axis=1)

# split by features (salaries)
X = df_encoded.drop(['EARN_MDN_1YR'], axis=1)
y = df_encoded['EARN_MDN_1YR']
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# train linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

#credlevels (degrees)
credlevs = [1, 2, 3, 4, 5, 6, 7, 8]

# table output
predicted_earnings = PrettyTable()
predicted_earnings.title = "Predicted Earnings 1 Year after Graduating for All Majors and CREDLEVs"
predicted_earnings.field_names = ["Major"] + [f"Predicted Earnings CREDLEV {credlev}" for credlev in credlevs]

for major in df['CIPDESC'].unique():
    cipcode = df[df['CIPDESC'] == major]['CIPCODE'].mean()
    row_data = [major]

    for credlev in credlevs:
        new_data = pd.DataFrame({
            'CIPDESC': [major],
            'CIPCODE': [cipcode],
            'CREDLEV': [credlev],
            'EARN_MDN_4YR': [np.nan], 
        })

        # encode on major
        new_data_encoded = pd.concat([new_data, pd.DataFrame(encoder.transform(new_data[['CIPDESC']]), columns=categories)], axis=1)
        new_data_encoded = new_data_encoded.drop('CIPDESC', axis=1)
        new_data_imputed = pd.DataFrame(imputer.transform(new_data_encoded), columns=new_data_encoded.columns)

        #predict using the trained model for the new data
        new_prediction = model.predict(new_data_imputed)
        row_data.append(f"${new_prediction[0]:,.2f}")
        
    predicted_earnings.add_row(row_data)

print(predicted_earnings)
```
